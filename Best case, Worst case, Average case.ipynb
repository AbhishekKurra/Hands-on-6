```python
import random
import time
import matplotlib.pyplot as plt

def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

def generate_best_case(n):
    return list(range(n))

def generate_worst_case(n):
    return list(range(n, 0, -1))

def generate_average_case(n):
    return [random.randint(0, n) for _ in range(n)]

def benchmark_quicksort():
    sizes = [100, 500, 1000, 5000, 10000, 20000]
    best_times, worst_times, avg_times = [], [], []

    for size in sizes:
        best_case = generate_best_case(size)
        worst_case = generate_worst_case(size)
        avg_case = generate_average_case(size)

        start = time.time()
        quicksort(best_case)
        best_times.append(time.time() - start)

        start = time.time()
        quicksort(worst_case)
        worst_times.append(time.time() - start)

        start = time.time()
        quicksort(avg_case)
        avg_times.append(time.time() - start)

    plt.plot(sizes, best_times, label='Best Case', marker='o')
    plt.plot(sizes, worst_times, label='Worst Case', marker='o')
    plt.plot(sizes, avg_times, label='Average Case', marker='o')

    plt.xlabel('Input Size (n)')
    plt.ylabel('Time (seconds)')
    plt.title('Quicksort Performance')
    plt.legend()
    plt.show()

data = [8, 4, 3, 1, 6, 7, 10]
print("Standard QuickSort:", quicksort(data))

benchmark_quicksort()
```
